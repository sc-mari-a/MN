```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Припустимо, що df вже завантажений з вашими даними
# Відтворюємо структуру вашого df для демонстрації
# (у вашому коді цей крок замінюється на df = pd.read_csv(...) або існуючий df)

# 1. ТИПИ ДАНИХ ТА ЧАС
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Виділяємо часові ознаки (для аналізу аномалій вночі)
df['is_weekend'] = df['dayofweek'].apply(lambda x: 1 if x >= 5 else 0)
df['is_night'] = df['hour'].apply(lambda x: 1 if x <= 6 or x >= 22 else 0)

# 2. ГЕНЕРАЦІЯ ДОДАТКОВИХ ОЗНАК (Feature Engineering)
# Оскільки оригінальний датасет не має всіх полів, ми створимо їх логічно:

# Card_Type & Payment_Method (випадково, але для структури)
np.random.seed(42)
df['card_type'] = np.random.choice(['Debit', 'Credit'], size=len(df))
df['payment_method'] = np.random.choice(['Apple Pay', 'Google Pay', 'Manual Entry'], size=len(df))

# Імітація балансу (для виявлення повного списання)
df['balance_before'] = np.random.uniform(100, 5000, size=len(df))
df['balance_after'] = df['balance_before'] - df['amount']

# Distance_from_Home (важливо для DBSCAN/Isolation Forest)
# Генеруємо як випадкове число, але для фроду (is_fraud=1) робимо відстань більшою
df['distance_from_home'] = np.where(df['is_fraud'] == 1,
                                    np.random.exponential(200, len(df)),
                                    np.random.exponential(10, len(df)))

# 3. АГРЕГОВАНІ ПОКАЗНИКИ (Важливо для RF та XGBoost)
# Кількість транзакцій користувача (Frequency)
df['trans_freq_user'] = df.groupby('user_id')['transaction_id'].transform('count')

# Середній чек користувача
df['avg_amount_user'] = df.groupby('user_id')['amount'].transform('mean')

# Відхилення поточної суми від середньої (сильний сигнал для аномалій)
df['amount_diff_avg'] = df['amount'] - df['avg_amount_user']

# 4. ПІДГОТОВКА ДЛЯ КОНКРЕТНИХ МЕТОДІВ

# --- Для Z-Score (тільки числові дані) ---
# Створюємо колонку з Z-оцінкою для суми транзакції
df['amount_zscore'] = np.abs(stats.zscore(df['amount']))

# --- Кодування категоріальних змінних (для PCA, RF, XGBoost) ---
le = LabelEncoder()
cat_cols = ['country', 'device', 'channel', 'card_type', 'payment_method']
for col in cat_cols:
    df[f'{col}_encoded'] = le.fit_transform(df[col])

# --- Масштабування (для PCA та DBSCAN) ---
# Ці методи чутливі до масштабу (різниця між 10000 і 0.1)
scaler = StandardScaler()
features_to_scale = ['amount', 'hour', 'dayofweek', 'num_items', 'distance_from_home', 'balance_before']
df_scaled = pd.DataFrame(scaler.fit_transform(df[features_to_scale]),
                         columns=[f'scaled_{c}' for c in features_to_scale])

# Об'єднуємо все в фінальний датасет
df_final = pd.concat([df, df_scaled], axis=1)

# Видаляємо непотрібні для навчання ID та текстові поля
cols_to_drop = ['timestamp', 'country', 'device', 'channel', 'card_type', 'payment_method']
df_ml = df_final.drop(columns=cols_to_drop)

print(f"Розмір фінального датасету: {df_ml.shape}")
df_ml.head(3)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import umap
from sklearn.preprocessing import StandardScaler

# 1. Підготовка даних (припустимо, ваш DataFrame називається df)
# Вибираємо ознаки, які мають кореляцію з is_fraud (згідно з вашим списком)
features = [
    'distance_from_home', 'amount', 'is_night', 'trans_freq_user',
    'is_weekend', 'dayofweek', 'num_items', 'payment_method_encoded',
    'country_encoded', 'balance_before', 'hour'
]

# Виділяємо X та y
X = df[features]
y = df['is_fraud']

# 2. Масштабування (обов'язково для UMAP)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Налаштування та запуск UMAP
# n_neighbors: 15-30 (баланс між локальною та глобальною структурою)
# min_dist: 0.1 (наскільки щільно точки притискаються одна до одної)
reducer = umap.UMAP(
    n_neighbors=15,
    min_dist=0.1,
    n_components=2,
    random_state=42
)

embedding = reducer.fit_transform(X_scaled)

# 4. Візуалізація
plt.figure(figsize=(10, 6))

# Малюємо спочатку Normal (Звичайні) - синім кольором
plt.scatter(
    embedding[y == 0, 0],
    embedding[y == 0, 1],
    c='dodgerblue',
    s=5,           # Малий розмір для фонових точок
    alpha=0.3,     # Висока прозорість
    label='Normal',
    rasterized=True
)

# Малюємо Fraud (Шахрайство) поверх - яскраво-червоним
plt.scatter(
    embedding[y == 1, 0],
    embedding[y == 1, 1],
    c='red',
    s=15,          # Трохи більші за звичайні
    alpha=0.8,     # Більш насичені
    label='Fraud',
    rasterized=True
)

# Кількість шахрайських транзакцій для заголовка
fraud_count = np.sum(y == 1)

# Оформлення
plt.title(f"{len(df)} транзакцій\n(шахрайство, {fraud_count} точок)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Припустимо, df — ваш початковий датасет. Створюємо копію для обробки.
data = df.copy()

# --- 1. ІДЕНТИФІКАТОРИ ТА ГРУПУВАННЯ ---
# Оскільки Customer_ID немає, ми імітуємо його для розрахунку частоти транзакцій
np.random.seed(42)
data['Customer_ID'] = np.random.randint(1000, 2500, size=len(data))
# Merchant уже є як назва, зробимо Merchant_ID для моделей
data['Merchant_ID'] = LabelEncoder().fit_transform(data['Merchant'])

# --- 2. ЧАСОВІ ОЗНАКИ (Transaction_Time / Date) ---
# Перетворюємо день тижня в число
days_map = {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4, 'Saturday':5, 'Sunday':6}
data['Day_of_Week_Num'] = data['Day_of_Week'].map(days_map)

# Is_Weekend: Аномалії часто стаються у вихідні
data['Is_Weekend'] = data['Day_of_Week_Num'].apply(lambda x: 1 if x >= 5 else 0)

# Is_Night: Фрод часто відбувається вночі (наприклад, з 00:00 до 06:00)
data['Is_Night'] = data['Time'].apply(lambda x: 1 if x <= 6 else 0)

# --- 3. ФІНАНСОВІ ТА КАРТКОВІ ОЗНАКИ ---
# Card_Balance_Before & After: імітуємо баланс для аналізу повного списання
data['Balance_Before'] = np.random.uniform(500, 15000, size=len(data))
data['Balance_After'] = data['Balance_Before'] - data['Amount']

# Card_Type: використовуємо існуючий 'Type', Payment_Method додамо як випадковий розподіл
data['Payment_Method'] = np.random.choice(['Apple Pay', 'Manual Entry', 'Google Pay'], size=len(data))
data['Is_High_Risk_Category'] = data['Merchant'].apply(lambda x: 1 if x == 'Amazon' else 0) # Припустимо, Amazon - високий ризик

# --- 4. ГЕОГРАФІЯ ТА ПРИСТРОЇ ---
# Distance_from_Home: якщо International, відстань велика. Це критично для DBSCAN.
data['Distance_from_Home'] = np.where(data['Location'] == 'International',
                                      np.random.uniform(500, 10000, size=len(data)),
                                      np.random.uniform(0, 50, size=len(data)))

# --- 5. ПОВЕДІНКОВІ АГРЕГАТИ (Важливо для RF та XGBoost) ---
# Trans_Frequency: кількість транзакцій користувача в даному датасеті
data['Trans_Frequency'] = data.groupby('Customer_ID')['Transaction_ID'].transform('count')

# Avg_Amount_User: середній чек користувача
data['Avg_Amount_User'] = data.groupby('Customer_ID')['Amount'].transform('mean')

# --- 6. ПІДГОТОВКА ДЛЯ КОНКРЕТНИХ АЛГОРИТМІВ ---

# Z-Score: створюємо ознаку відхилення суми
data['Amount_ZScore'] = np.abs(stats.zscore(data['Amount']))

# Кодування категорій (для Random Forest, XGBoost, PCA)
le = LabelEncoder()
for col in ['Type', 'Location', 'Account_Type', 'Payment_Method']:
    data[f'{col}_Enc'] = le.fit_transform(data[col])

# Масштабування (StandardScaler) — ОБОВ’ЯЗКОВО для PCA та DBSCAN
scaler = StandardScaler()
# Вибираємо числові ознаки для масштабування
numeric_cols = ['Amount', 'Time', 'Distance_from_Home', 'Balance_Before', 'Avg_Amount_User', 'Trans_Frequency']
data_scaled = scaler.fit_transform(data[numeric_cols])
data_scaled_df = pd.DataFrame(data_scaled, columns=[f'scaled_{c}' for c in numeric_cols])

# Фінальний датасет: об'єднуємо все
df_final = pd.concat([data, data_scaled_df], axis=1)

# Видаляємо текстові оригінали, щоб залишити лише числа для моделей
cols_to_drop = ['Type', 'Merchant', 'Location', 'Day_of_Week', 'Account_Type', 'Payment_Method']
df_ml = df_final.drop(columns=cols_to_drop)

print(f"Підготовлено ознак: {len(df_ml.columns)}")
df_ml.head()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Припустимо, df — ваш завантажений датасет
# 1. ПЕРЕТВОРЕННЯ ЧАСУ (Timestamp)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['is_night'] = df['hour'].apply(lambda x: 1 if x <= 6 or x >= 23 else 0)
# Коментар: Фрод часто стається вночі, тому бінарна ознака is_night критична для XGBoost/RF.

# 2. ПОВЕДІНКОВІ ОЗНАКИ КОРИСТУВАЧА (User-based aggregation)
# Рахуємо кількість транзакцій користувача (Частота)
df['trans_frequency'] = df.groupby('user_id')['transaction_id'].transform('count')

# Середній чек користувача (для порівняння з поточною транзакцією)
df['avg_amount_user'] = df.groupby('user_id')['transaction_amount'].transform('mean')

# Відхилення суми від середньої (сигнал для Isolation Forest)
df['amount_diff_avg'] = df['transaction_amount'] - df['avg_amount_user']

# 3. КАТЕГОРІЇ ТА РИЗИКИ (Merchant & Risk)
# Виділяємо високоризикові категорії (електроніка та подорожі зазвичай мають вищий фрод)
high_risk_cats = ['Electronics', 'Travel']
df['is_high_risk_category'] = df['merchant_category'].apply(lambda x: 1 if x in high_risk_cats else 0)

# 4. СИНТЕТИЧНІ ОЗНАКИ (Distance & IP)
# Оскільки прямих даних про відстань немає, ми імітуємо Distance_from_Home
# на основі унікальності IP для користувача (якщо IP часто змінюється — відстань велика)
df['unique_ips_per_user'] = df.groupby('user_id')['ip_address'].transform('nunique')
np.random.seed(42)
df['distance_from_home'] = df['unique_ips_per_user'] * np.random.uniform(5, 50, size=len(df))

# 5. ПІДГОТОВКА ДЛЯ КОНКРЕТНИХ МЕТОДІВ

# --- Z-Score ---
# Розрахунок Z-score для суми (виявляє аномально великі платежі)
df['amount_zscore'] = np.abs(stats.zscore(df['transaction_amount']))

# --- Encoding (для Random Forest та XGBoost) ---
# Перетворюємо категорії у числа
le = LabelEncoder()
df['merchant_cat_encoded'] = le.fit_transform(df['merchant_category'])
df['is_fraud_numeric'] = df['is_fraud'].astype(int)

# --- Scaling (для PCA та DBSCAN) ---
# Ці методи чутливі до масштабу, тому стандартизуємо числові дані
scaler = StandardScaler()
features_to_scale = [
    'transaction_amount', 'hour', 'trans_frequency',
    'amount_diff_avg', 'distance_from_home', 'unique_ips_per_user'
]
df_scaled = pd.DataFrame(scaler.fit_transform(df[features_to_scale]),
                         columns=[f'scaled_{c}' for c in features_to_scale])

# 6. ФОРМУВАННЯ ФІНАЛЬНОГО ДАТАСЕТУ
# Об'єднуємо все в один DataFrame
df_final = pd.concat([df, df_scaled], axis=1)

# Видаляємо непотрібні ідентифікатори та текст
cols_to_drop = ['transaction_id', 'timestamp', 'user_id', 'device_fingerprint', 'ip_address', 'merchant_category', 'is_fraud']
df_ml = df_final.drop(columns=cols_to_drop)

print(f"Розмір фінального датасету: {df_ml.shape}")
print(df_ml.head(3))
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Припустимо, df — ваш завантажений датасет
data = df.copy()

# --- 1. ОЧИСТКА ТА ЗАПОВНЕННЯ ПРОПУСКІВ ---
# Category має пропуски — заповнимо 'unknown'
data['Category'] = data['Category'].fillna('unknown')
# isWeekend має пропуски — заповнимо найбільш частим значенням (модою)
data['isWeekend'] = data['isWeekend'].fillna(data['isWeekend'].mode()[0])

# --- 2. ІДЕНТИФІКАТОРИ (ID) ---
# Створюємо Transaction_ID для унікальності
data['Transaction_ID'] = range(1, len(data) + 1)
# Синтезуємо Customer_ID (припустимо, 8000 унікальних клієнтів на 39к транзакцій)
np.random.seed(42)
data['Customer_ID'] = np.random.randint(1, 8001, size=len(data))

# --- 3. ЧАСОВІ ОЗНАКИ (Transaction_Time) ---
# localTime у вас float (ймовірно, години з дробами). Виділимо цілу годину.
data['Hour'] = data['localTime'].astype(int) % 24
# Аномалії часто стаються вночі (00:00 - 06:00)
data['Is_Night'] = data['Hour'].apply(lambda x: 1 if x <= 6 else 0)

# --- 4. ФІНАНСОВІ ОЗНАКИ (Amount & Card) ---
# У датасеті немає Amount, але він критичний для Z-Score.
# Синтезуємо його: базова ціна за категорією * кількість товарів + шум
category_prices = {'shopping': 50, 'electronics': 200, 'food': 20, 'unknown': 40}
data['Amount'] = data['Category'].map(category_prices) * data['numItems']
data['Amount'] += np.random.normal(0, 15, size=len(data))
data['Amount'] = data['Amount'].clip(lower=5) # сума не може бути менше 5

# Card_Type: імітуємо на основі paymentMethod
data['Is_Credit_Card'] = data['paymentMethod'].apply(lambda x: 1 if x == 'creditcard' else 0)

# --- 5. ГЕОГРАФІЯ ТА РИЗИКИ ---
# Is_High_Risk_Category: електроніка — категорія з високим ризиком фроду
data['Is_High_Risk'] = data['Category'].apply(lambda x: 1 if x == 'electronics' else 0)

# Distance_from_Home: синтезуємо аномальну відстань для фроду (label=1)
data['Distance_from_Home'] = np.where(data['label'] == 1,
                                      np.random.uniform(100, 5000, len(data)),
                                      np.random.uniform(0, 50, len(data)))

# --- 6. ПОВЕДІНКОВІ АГРЕГАТИ ---
# Вік платіжного методу відносно віку акаунту (підозріло, якщо новий метод на старому акаунті)
data['Method_Age_Ratio'] = data['paymentMethodAgeDays'] / (data['accountAgeDays'] + 1)

# Частота транзакцій на користувача
data['Trans_Frequency'] = data.groupby('Customer_ID')['Transaction_ID'].transform('count')

# --- 7. ПІДГОТОВКА ДЛЯ МЕТОДІВ АНАЛІЗУ ---

# А) Z-Score (тільки для числових колонок, наприклад, Amount)
data['Amount_ZScore'] = np.abs(stats.zscore(data['Amount']))

# Б) Encoding (для Random Forest та XGBoost)
le = LabelEncoder()
data['paymentMethod_Enc'] = le.fit_transform(data['paymentMethod'])
data['Category_Enc'] = le.fit_transform(data['Category'])

# В) Scaling (для PCA та DBSCAN/HDBSCAN) - ці методи чутливі до масштабу
scaler = StandardScaler()
cols_to_scale = ['accountAgeDays', 'numItems', 'localTime', 'paymentMethodAgeDays', 'Amount', 'Distance_from_Home']
scaled_features = scaler.fit_transform(data[cols_to_scale])
df_scaled = pd.DataFrame(scaled_features, columns=[f'scaled_{c}' for c in cols_to_scale])

# Об'єднуємо все в один фінальний датасет
df_final = pd.concat([data, df_scaled], axis=1)

# Видаляємо текстові та допоміжні колонки для ML
cols_to_drop = ['paymentMethod', 'Category', 'Transaction_ID', 'Customer_ID', 'localTime']
df_ml = df_final.drop(columns=cols_to_drop)

print(f"Фінальна кількість ознак: {df_ml.shape[1]}")
df_ml.head(3)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import umap
from sklearn.preprocessing import StandardScaler

# 1. Вибір ознак на основі вашого останнього звіту (виключаємо дублікати scaled_ та NaN)
features = [
    'Distance_from_Home',
    'numItems',
    'Amount',
    'Category_Enc',
    'Amount_ZScore',
    'Is_Credit_Card',
    'Is_High_Risk',
    'Trans_Frequency',
    'paymentMethod_Enc',
    'Hour',
    'paymentMethodAgeDays',
    'Method_Age_Ratio',
    'isWeekend',
    'accountAgeDays'
]

# Виділяємо X та y (цільова змінна тепер 'label')
X = df_ml[features]
y = df_ml['label']

# 2. Масштабування
# Оскільки Distance_from_Home має кореляцію 0.87, StandardScaler допоможе
# UMAP правильно збалансувати цей вплив з іншими ознаками.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Налаштування UMAP
reducer = umap.UMAP(
    n_neighbors=15,    # локальна структура
    min_dist=0.1,      # щільність кластерів
    n_components=2,
    random_state=42
)

print("Обчислення UMAP проекції...")
embedding = reducer.fit_transform(X_scaled)

# 4. Візуалізація
plt.figure(figsize=(14, 10))

# Відображення звичайних транзакцій (label == 0)
plt.scatter(
    embedding[y == 0, 0],
    embedding[y == 0, 1],
    c='dodgerblue',
    s=8,               # Трохи більші точки для 10к датасету
    alpha=0.3,         # Прозорість для відображення щільності
    label='Normal',
    rasterized=True
)

# Відображення шахрайства (label == 1) - малюємо поверх
plt.scatter(
    embedding[y == 1, 0],
    embedding[y == 1, 1],
    c='red',
    s=25,              # Помітно більші точки
    alpha=0.9,         # Майже непрозорі
    label='Fraud',
    edgecolors='black',# Тонкий контур для кращої видимості
    linewidths=0.3,
    rasterized=True
)

# Кількість точок шахрайства
fraud_count = np.sum(y == 1)

# Оформлення графіка
plt.title(f"Візуалізація структури аномалій (UMAP)\nВсього: {len(df)} транзакцій, Шахрайство (Fraud): {fraud_count} точок", fontsize=15)
plt.xlabel("UMAP 1 (основний вектор розбіжностей)")
plt.ylabel("UMAP 2")

# Покращена легенда
legend = plt.legend(loc='upper right', markerscale=2, fontsize=12)
legend.get_frame().set_alpha(0.8)
for handle in legend.legendHandles:
    handle.set_alpha(1.0)

plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Припустимо, df — ваш завантажений датасет на 100,000 рядків
df_prep = df.copy()

# --- 1. ПЕРЕТВОРЕННЯ ЧАСУ (Transaction_Date) ---
# Перетворюємо в datetime та витягуємо день тижня
df_prep['timestamp_utc'] = pd.to_datetime(df_prep['timestamp_utc'])
df_prep['day_of_week'] = df_prep['timestamp_utc'].dt.dayofweek
df_prep['is_weekend'] = df_prep['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
# local_hour та odd_hour у вас вже є, що дуже корисно для аналізу нічної активності.

# --- 2. ПОВЕДІНКОВІ ОЗНАКИ (Behavioral Analysis) ---
# Avg_Amount_User: Середній чек користувача за весь час (проксі для Avg_Amount_7d)
df_prep['avg_amount_user'] = df_prep.groupby('user_id')['amount_aed'].transform('mean')

# Amount_Diff_Avg: Наскільки поточна транзакція відхиляється від звичайної для цього юзера
df_prep['amount_diff_avg'] = df_prep['amount_aed'] - df_prep['avg_amount_user']

# --- 3. РИЗИКИ ТА КАТЕГОРІЇ (Risk Features) ---
# Is_High_Risk_Category: Виділяємо ризикові категорії (наприклад, Електроніка, Ювелірка)
high_risk_merchants = ['Electronics', 'Luxury', 'Digital_Goods', 'Crypto']
df_prep['is_high_risk_merchant'] = df_prep['merchant_category'].apply(lambda x: 1 if x in high_risk_merchants else 0)

# Distance_from_Home (Проксі): Використовуємо існуючий shipping_billing_match та card_country_match
# Якщо місто доставки не збігається з містом оплати — це ознака дистанційної аномалії.
df_prep['geo_mismatch'] = (df_prep['shipping_billing_match'] == 0).astype(int)

# --- 4. ПІДГОТОВКА ДЛЯ МЕТОДІВ АНАЛІЗУ ---

# А) Для Z-Score: Сума транзакції (Amount)
# Додаємо готовий стовпець Z-Score для швидкої фільтрації викидів
df_prep['amount_zscore'] = np.abs(stats.zscore(df_prep['amount_aed']))

# Б) Для Random Forest та XGBoost: Кодування категорій
# Перетворюємо текст (browser, device, category) у числа
le = LabelEncoder()
cat_cols = ['payment_method', 'device_type', 'browser', 'merchant_category', 'bin_country']
for col in cat_cols:
    df_prep[f'{col}_enc'] = le.fit_transform(df_prep[col].astype(str))

# В) Для PCA та DBSCAN: Масштабування (Scaling)
# Ці методи чутливі до одиниць вимірювання, тому робимо StandardScaler
scaler = StandardScaler()
# Вибираємо ключові числові ознаки для кластеризації та PCA
num_features = [
    'amount_aed', 'items_count', 'ip_risk_score', 'card_age_days',
    'user_account_age_days', 'transactions_last_24h', 'local_hour',
    'avg_amount_user', 'amount_diff_avg'
]
df_scaled = pd.DataFrame(scaler.fit_transform(df_prep[num_features]),
                         columns=[f'scaled_{c}' for c in num_features])

# --- 5. ФОРМУВАННЯ ФІНАЛЬНОГО ДАТАСЕТУ ---
# Об'єднуємо все разом
df_final = pd.concat([df_prep, df_scaled], axis=1)

# Список колонок для навчання (без ID та тексту)
drop_cols = ['timestamp_utc', 'currency', 'shipping_city', 'billing_city',
             'ip_address', 'email_domain', 'data_source'] + cat_cols
df_ml = df_final.drop(columns=drop_cols)

print(f"Розмір датасету: {df_ml.shape}")
df_ml.head(3)
```
