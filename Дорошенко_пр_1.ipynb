**Завдання 15**

5. Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить найбільший елемент у кожному рядку матриці.

```python
m15 = np.random.rand(7,7)
v15 = np.max(m15, axis=1).reshape(-1,1) # axis=1 - для пошуку в рядку, і змінюємо розмір на вертекальний(стовпчик)
print("Матриця:\n", m15)
print("\nВектор-стовпчик:\n", v15)
```

# Висновок

В даній лабораторній роботі було використано методи, з бібліотеки numpy, для створення і роботи з матрицями та векторами. Використовувалися такі функції:
- np.arange – послідовність(вектор, якщо не змінювати розмір) чисел із заданим початком, кінцем та кроком

- np.random – генерації випадкових чисел (цілих або дійсних)

- .reshape – змінює форму

- .T – транспонування

- знаки +, -, * – виконують операції між масивами по елементам

- np.dot – множення матриць

- np.prod – добуток усіх елементів масиву

- np.max – максимальне значення у масиві (можна по всьому масиву чи по рядках(axis=1)/стовпцях(axis=0))

- np.linalg.inv – обернена матриця
